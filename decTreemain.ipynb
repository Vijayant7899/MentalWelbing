# Preprocessing

```python
# imports
import pandas as pd
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv('survey1.csv')
df
```

```python
# dropping insignificant columns
df = df.drop(['Please write a very short summary of your lockdown experience in terms of your physical, mental, and emotional health.', "Editor's note: We hope you liked our questionnaire; Let us know how much.", "Unnamed: 29", "Unnamed: 32", "Unnamed: 0", "Remarks (section wise rating: I+ II+ III+ IV+ V)"], axis = 1)

# using 2nd row as header
df.loc[0, 'Output Level (1-5)'] = "output"
headers = df.iloc[0]
df  = pd.DataFrame(df.values[1:], columns=headers)

# display dataframe
df.head()
```

```python
# checking count of Nan values in each column
df.isnull().sum()
```

```python
# replacing Nan values
df["Q25a"].fillna("NA", inplace = True)
df["Q25b"].fillna("NA", inplace = True)
```

```python
# df["Q25 (i)"].fillna("NA", inplace = True)
# df["Q25 (ii)"].fillna("NA", inplace = True)
```

```python
df.head()
```

```python
# correcting output values
for index, row in df.iterrows():
    row["output"] = row["output"][-1]
df.head(10)
```

```python
Y = df['output']
df = df.drop(['output'], axis = 1)
df.head()
```

```python
# converting labels to numbers

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in df.columns:
    if col not in ['Q9', 'Q20', 'output']:
        df[col]= le.fit_transform(df[col])

df.head(10)
```

```python
# joblib.dump(le, 'label_encoder.pkl')
```

```python
# ['Studying /being academically inclined', 'Household chores', 'Netflix or Other OTT', 'Sleeping', 'Gaming']
df['Q9(1)'] = 0
df['Q9(2)'] = 0
df['Q9(3)'] = 0
df['Q9(4)'] = 0
df['Q9(5)'] = 0
for index, row in df.iterrows():
    items = row['Q9'].split(",")
    for i in range(len(items)):
        items[i] = items[i].strip()
    for item in items:
        if item=="Studying /being academically inclined": df.loc[index, 'Q9(1)'] = 1
        elif item=="Household chores": df.loc[index, 'Q9(2)'] = 1
        elif item=="Netflix or Other OTT": df.loc[index, 'Q9(3)'] = 1
        elif item=="Sleeping": df.loc[index, 'Q9(4)'] = 1
        else: df.loc[index, 'Q9(5)'] = 1

df = df.drop(['Q9'], axis = 1)
```

```python
# ['Moody', 'Satisfied', 'Agitated/ Sad', 'Happy', 'Sad', 'Agitated/ Irritable']
df['Q20(1)'] = 0
df['Q20(2)'] = 0
df['Q20(3)'] = 0
df['Q20(4)'] = 0
df['Q20(5)'] = 0
for index, row in df.iterrows():
    items = row['Q20'].split(",")
    for i in range(len(items)):
        items[i] = items[i].strip()
    for item in items:
        if item=="Moody": df.loc[index, 'Q20(1)'] = 1
        elif item=="Satisfied": df.loc[index, 'Q20(2)'] = 1
        elif item=="Happy": df.loc[index, 'Q20(3)'] = 1
        elif item=="Sad": df.loc[index, 'Q20(4)'] = 1
        else: df.loc[index, 'Q20(5)'] = 1

df = df.drop(['Q20'], axis = 1)
```

Normalizing our data is very important, as we want the input features to be on the same order of magnitude to make our training easier. We'll use a min-max scaler from scikit-learn which scales our data to be between 0 and 1.

```python
from sklearn import preprocessing
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
```

```python
X_scale
```

Lastly, we wish to set aside some parts of our dataset for a validation set and a test set. We use the function train_test_split from scikit-learn to do that.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.3)
```

```python
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.7)
```

```python
print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf=clf.fit(X_train,Y_train)
```

```python
cy_pred = clf.predict(X_test)
cy_pred
cy_test=Y_test
```

```python
print("Accuracy:",metrics.accuracy_score(cy_test, cy_pred))
```
